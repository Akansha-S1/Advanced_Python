```python
#1
import matplotlib.pyplot as plt
names =[]
marks =[]
colors=['yellow','blue','green']
f = open('details.txt','r')
for row in f: 
    row = row.split(' ')
    names.append(row[0])
    marks.append(int(row[1]))
plt.bar(names,marks,color ='g',label='File Data')
plt.xlabel('Student names',fontsize =12)
plt.ylabel('Marks',fontsize=12)
plt.title('Student Marks',fontsize=20) 
plt.legend()
plt.show()
plt.pie(marks,labels=names,autopct='%1.1f%%',colors=colors,startangle=90,shadow=True)
plt.show()
```
