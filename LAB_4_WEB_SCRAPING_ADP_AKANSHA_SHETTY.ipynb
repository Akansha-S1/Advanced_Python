```python
import requests
res=requests.get('https://webscraping.com')
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(res.content, 'html.parser')
print(soup.prettify())
```

```python
for element in soup.select('div'):
    val=element.find('div',{"class","text"})
    print(val)
```

```python
for element in soup.select('div'):
    val=element.find('div',{"class","mb-srp__card__"})
    if val=="Recieve your data":
        print(val)
```
